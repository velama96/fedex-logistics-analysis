```
import pandas as pd

# Load datasets
combined = pd.read_csv('../data/combined_logistics_data.csv')
routes = pd.read_csv('../data/route_optimization_data.csv')
drivers = pd.read_csv('../data/driver_behavior_data.csv')
vehicles = pd.read_csv('../data/vehicle_data.csv')
fuel = pd.read_csv('../data/fuel_consumption_data.csv')
feedback = pd.read_csv('../data/customer_feedback_data.csv')
maintenance = pd.read_csv('../data/maintenance_alerts_data.csv')
env = pd.read_csv('../data/environmental_data.csv')
inventory = pd.read_csv('../data/inventory_data.csv')

datasets = {
    'Combined Logistics': combined,
    'Routes': routes,
    'Drivers': drivers,
    'Vehicles': vehicles,
    'Fuel': fuel,
    'Customer Feedback': feedback,
    'Maintenance Alerts': maintenance,
    'Environmental': env,
    'Inventory': inventory
}

for name, df in datasets.items():
    print(f"--- {name} ---")
    print(df.shape)
    print(df.head(), "\n")
```
